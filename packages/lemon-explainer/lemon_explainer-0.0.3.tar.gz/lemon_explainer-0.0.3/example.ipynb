```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

# Load dataset
data = load_iris(as_frame=True)
X = data.data
y = pd.Series(np.array(data.target_names)[data.target])
y.name = "Class"
```

```python
from sklearn.ensemble import RandomForestClassifier

# Train complex model
clf = RandomForestClassifier(random_state=4)
clf.fit(X, y)
```

```python
from lemon import LemonExplainer

# Instantiate explainer
explainer = LemonExplainer(X, random_state=4)


# Choose and explain prediction for an instance
instance = X.iloc[-1, :]
exp = explainer.explain_instance(instance, clf.predict_proba)[0]
exp.show_in_notebook()
```
